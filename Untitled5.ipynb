```python
import matplotlib.animation as animation
import numpy as np
from pylab import *
import netCDF4 as nc
import glob
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.backends.backend_pdf
from mpl_toolkits.axes_grid1 import AxesGrid
import warnings
import multiprocessing
import ctypes as c
from datetime import datetime, time, timedelta
from scipy.interpolate import interp1d
from optparse import OptionParser
from tqdm import tnrange, notebook
from time import sleep
import xarray as xr
from contextlib import contextmanager
from scipy.spatial import cKDTree
from functools import partial
import time
import os
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from skimage.measure import label as sklabel

global fF, z_interp, fsW, fsC, cbar_set
Rd = 287.
Cp = 1004.
g = 9.81

@contextmanager
def poolcontext(*args, **kwargs):
    pool = multiprocessing.Pool(*args, **kwargs)
    yield pool
    pool.terminate()
    
def interp3dz_np(j,data, z3d, z1d):
    
    dinterp = np.interp(z1d, z3d[:,j], data[:,j])

    return dinterp

def getzw_fv3(t):
    print('FV3')
    zF = np.array(-np.cumsum(fF.delz[t,::-1,cross_y,xlim],axis=0))

    if (plot_var=='W'):
        wF = np.array(fF.dzdt[t,::-1,cross_y,xlim])
    elif (plot_var=='U'):
        uF = np.array(fF.ugrd[t,::-1,cross_y,xlim])
        wpre = np.array(fF.dzdt[t,::-1,cross_y,xlim])
        wF = np.stack((uF,wpre),axis=0)
    elif (plot_var=='$\Theta$'):
        pres = np.array(fF.pres[0,::-1,cross_y,xlim])
        vap = np.array(fF.spfh[0,::-1,cross_y,xlim])
        pres = pres / (1 + vap * 0.622)
        wF = np.array(fF.tmp[t,::-1,cross_y,xlim]) * (100000.0/pres)**(Rd/Cp)
        wpre = np.array(fF.dzdt[t,::-1,cross_y,xlim])
        wF = np.stack((wF,wpre),axis=0)
    elif (plot_var=='P\''):
        pres = np.array(fF.pres[0,::-1,cross_y,xlim])
        vap = np.array(fF.spfh[0,::-1,cross_y,xlim])
        pres = pres / (1 + vap * 0.622)
        wpre = np.array(fF.dzdt[t,::-1,cross_y,xlim])
        wF = np.stack((pres,wpre),axis=0)
    elif (plot_var=='dz\''):
        dz = np.array(-fF.delz[t,::-1,cross_y,xlim])
        wpre = np.array(fF.dzdt[t,::-1,cross_y,xlim])
        wF = np.stack((dz,wpre),axis=0)
    return zF, wF

def getzw_wrf(t):
    print('WRF')
    filenameW = fsW[t]
    fW = xr.open_dataset(filenameW ,engine='netcdf4')
    zW = (fW.PH[0,:,cross_y,xlim]+fW.PHB[0,:,cross_y,xlim])/9.81 - fW.HGT[0,cross_y,xlim]
    
    if (not plot_var == 'W'):
        zW = 0.5 * (zW[:-1,:]+zW[1:,:])
        
    if (plot_var=='W'):
        wW= fW.W[0,:,cross_y,xlim]
    elif (plot_var=='U'):
        xlim_cur = np.append(xlim,xlim[-1]+1)
        uW= fW.U[0,:,cross_y,xlim_cur]
        uW = 0.5 * (uW[:,:-1]+uW[:,1:])
        wpre = fW.W[0,:,cross_y,xlim]
        wpre = 0.5 * (wpre[:-1,:]+wpre[1:,:])
        wW = np.stack((uW,wpre),axis=0)
    elif (plot_var=='$\Theta$'):
        uW = np.array(fW.T[0,:,cross_y,xlim]+300.0)
        wpre = fW.W[0,:,cross_y,xlim]
        wpre = 0.5 * (wpre[:-1,:]+wpre[1:,:])
        wW = np.stack((uW,wpre),axis=0)
    elif (plot_var=='P\''):
        uW = np.array(fW.P[0,:,cross_y,xlim]+fW.PB[0,:,cross_y,xlim])
        wpre = fW.W[0,:,cross_y,xlim]
        wpre = 0.5 * (wpre[:-1,:]+wpre[1:,:])
        wW = np.stack((uW,wpre),axis=0)
    elif (plot_var=='dz\''):
        z = np.array(fW.PHB[0,:,cross_y,xlim] + fW.PH[0,:,cross_y,xlim])/9.81
        dz = z[1:,:]-z[:-1,:]
        wpre = fW.W[0,:,cross_y,xlim]
        wpre = 0.5 * (wpre[:-1,:]+wpre[1:,:])
        wW = np.stack((dz,wpre),axis=0)
    return zW,wW

def getzw_cm1(t):
    print('CM1')
    filename = fsC[t]
    f = xr.open_dataset(filename ,engine='netcdf4')
    
    if (plot_var=='W'):
        w = np.array(f.winterp[0,:,167,xlim])
        nz,nx = w.shape 
    elif (plot_var=='U'):
        xlim_cur = np.append(xlim,xlim[-1]+1)
        u = np.array(f.u[0,:,167,xlim_cur])
        nz,nx = u.shape 
        u = 0.5 * (u[:,:-1]+u[:,1:])
        w_pre = np.array(f.winterp[0,:,167,xlim])
        w = np.stack((u,w_pre),axis=0)
    elif (plot_var=='$\Theta$'):
        u = np.array(f.th[0,:,167,xlim])
        nz,nx = np.shape(u)
        w_pre = np.array(f.winterp[0,:,167,xlim])
        w = np.stack((u,w_pre),axis=0)
    elif (plot_var=='P\''):
        p = np.array(f.prs0[0,:,167,xlim]+f.prspert[0,:,167,xlim])
        nz,nx = np.shape(p)
        w_pre = np.array(f.winterp[0,:,167,xlim])
        w = np.stack((p,w_pre),axis=0)
    elif (plot_var=='dz\''):
        w_pre = np.array(f.winterp[0,:,167,xlim])
        nz,nx = np.shape(w_pre)
        dz = w_pre * 0.0
        w = np.stack((dz,w_pre),axis=0)
        
    z = np.broadcast_to(f.zh,(nx,nz)).transpose()*1000.0
    return z,w

def plot_cross(t):
    global cbar_set
    
    print("time = %i"%(t))
        
    zF,wF = getzw_fv3(t)
    zW,wW = getzw_wrf(t)
    zC,wC = getzw_cm1(t)
    
    with poolcontext(processes=12) as pool:
        #interp3dz_np(w[:,y,x], z[:,y,x], z_interp)
        if (plot_var=='W'):
            w_crossF = pool.map(partial(interp3dz_np,
                                 data=wF,z3d=zF,z1d=z_interp),np.arange(size(xlim)))
            w_crossW = pool.map(partial(interp3dz_np,
                                 data=wW,z3d=zW,z1d=z_interp),np.arange(size(xlim)))
            w_crossC = pool.map(partial(interp3dz_np,
                                 data=wC,z3d=zC,z1d=z_interp),np.arange(size(xlim)))
        elif (not plot_var == 'dz\''):
            w_crossF = pool.map(partial(interp3dz_np,
                                 data=wF[1,:],z3d=zF,z1d=z_interp),np.arange(size(xlim)))
            w_crossW = pool.map(partial(interp3dz_np,
                                 data=wW[1,:],z3d=zW,z1d=z_interp),np.arange(size(xlim)))
            w_crossC = pool.map(partial(interp3dz_np,
                                 data=wC[1,:],z3d=zC,z1d=z_interp),np.arange(size(xlim)))
            
            var_crossF = pool.map(partial(interp3dz_np,
                                 data=wF[0,:],z3d=zF,z1d=z_interp),np.arange(size(xlim)))
            var_crossW = pool.map(partial(interp3dz_np,
                                 data=wW[0,:],z3d=zW,z1d=z_interp),np.arange(size(xlim)))
            var_crossC = pool.map(partial(interp3dz_np,
                                 data=wC[0,:],z3d=zC,z1d=z_interp),np.arange(size(xlim)))
        else :
            w_crossF = pool.map(partial(interp3dz_np,
                                 data=wF[1,:],z3d=zF,z1d=z_interp),np.arange(size(xlim)))
            w_crossW = pool.map(partial(interp3dz_np,
                                 data=wW[1,:],z3d=zW,z1d=z_interp),np.arange(size(xlim)))
            w_crossC = pool.map(partial(interp3dz_np,
                                 data=wC[1,:],z3d=zC,z1d=z_interp),np.arange(size(xlim)))
            
            var_crossF = wF[0,:]
            var_crossW = wW[0,:]
            var_crossC = wC[0,:]

    if (plot_var=='P\''):
        var_crossF = var_crossF - p0F
        var_crossW = var_crossW - p0W
        var_crossC = var_crossC - p0C
    elif (plot_var=='dz\''):
        var_crossF = var_crossF - np.broadcast_to(dz0F,(var_crossF.shape[1],var_crossF.shape[0])).transpose()
        var_crossW = var_crossW - np.broadcast_to(dz0W,(var_crossW.shape[1],var_crossW.shape[0])).transpose()

        
    grid[0].clear()
    grid[1].clear()
    grid[2].clear()
    
    if plot_var == 'dz\'':
        w_contoursWf = grid[0].pcolormesh(np.arange(0,3.0*size(xlim),3.0),z0W,var_crossW,vmin = limits[0], vmax = limits[1], cmap=newcmp,shading='gouraud')
        w_contoursW = grid[0].contour(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(w_crossW),levels=con_levels,colors=('black'))

        w_contoursFf = grid[1].pcolormesh(np.arange(0,3.0*size(xlim),3.0),z0F,var_crossF,vmin = limits[0], vmax = limits[1], cmap=newcmp,shading='gouraud')
        w_contoursF = grid[1].contour(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(w_crossF),levels=con_levels,colors=('black'))

        w_contoursCf = grid[2].pcolormesh(np.arange(0,3.0*size(xlim),3.0),zC,var_crossC,vmin = limits[0], vmax = limits[1], cmap=newcmp,shading='gouraud')
        w_contoursC = grid[2].contour(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(w_crossC),levels=con_levels,colors=('black'))
    elif plot_var == 'W':
        w_contoursWf = grid[0].pcolormesh(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(w_crossW),vmin = limits[0], vmax = limits[1], cmap=newcmp,shading='gouraud')
        w_contoursW = grid[0].contour(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(w_crossW),levels=con_levels,colors=('black'))

        w_contoursFf = grid[1].pcolormesh(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(w_crossF),vmin = limits[0], vmax = limits[1], cmap=newcmp,shading='gouraud')
        w_contoursF = grid[1].contour(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(w_crossF),levels=con_levels,colors=('black'))

        w_contoursCf = grid[2].pcolormesh(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(w_crossC),vmin = limits[0], vmax = limits[1], cmap=newcmp,shading='gouraud')
        w_contoursC = grid[2].contour(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(w_crossC),levels=con_levels,colors=('black'))

    else:
        w_contoursWf = grid[0].pcolormesh(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(var_crossW),vmin = limits[0], vmax = limits[1], cmap=newcmp,shading='gouraud')
        w_contoursW = grid[0].contour(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(w_crossW),levels=con_levels,colors=('black'))

        w_contoursFf = grid[1].pcolormesh(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(var_crossF),vmin = limits[0], vmax = limits[1], cmap=newcmp,shading='gouraud')
        w_contoursF = grid[1].contour(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(w_crossF),levels=con_levels,colors=('black'))

        w_contoursCf = grid[2].pcolormesh(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(var_crossC),vmin = limits[0], vmax = limits[1], cmap=newcmp,shading='gouraud')
        w_contoursC = grid[2].contour(np.arange(0,3.0*size(xlim),3.0),z_interp,np.transpose(w_crossC),levels=con_levels,colors=('black'))



    if (not cbar_set) : 
        cbar = grid.cbar_axes[0].colorbar(w_contoursWf)
        cbar.set_label(label='%s (%s)'%(plot_var,var_units), fontsize=20)
        cbar.ax.tick_params(labelsize=20)
        cbar_set = True

    grid[0].set_xlabel("X (km)", fontsize=20)
    grid[0].set_title("WRF",fontsize=20)
    grid[0].tick_params(axis='x', labelsize=16) 
    grid[0].tick_params(axis='y', labelsize=16)
    grid[0].set_ylabel("Height (m)", fontsize=20)
    grid[1].tick_params(axis='x', labelsize=16) 
    grid[1].set_xlabel("X (km)", fontsize=20)
    grid[1].set_title("FV3",fontsize=20)
    grid[2].tick_params(axis='x', labelsize=16) 
    grid[2].set_xlabel("X (km)", fontsize=20)
    grid[2].set_title("CM1",fontsize=20)

    plt.title("Cross Section of %s at t = %i s "%(plot_var,(t+1)*25))    
    
    return cbar_set, w_contoursWf, w_contoursFf, w_contoursCf #,w_contoursW, w_contoursF, w_contoursC, cbar_set
    

# Open the NetCDF file
filenameF = "/oldscratch/larissa.reames/WoF_FV3SAR/20190520_periodic/201905202200/ENS01/c3500/fv3_history.nc"

fF = xr.open_dataset(filenameF ,engine='netcdf4')
cross_y = 166
xlim = np.arange(17,29)
time = 40
plot_var = 'W'
var_units = 'm $s^{-1}$'
limits=(-20,60)
#con_levels=np.append(np.arange(-20,0,2.5),np.arange(10,60,10))
con_levels = np.arange(-20,0,1)
arr = np.flip(np.append(np.linspace(0,0.5,126),np.linspace(0.5,1.0,42)))
cmap = plt.get_cmap('RdBu')
newcolors = cmap(arr)
newcmp = colors.ListedColormap(newcolors)
#newcmp = plt.get_cmap('turbo')
                     
string = "/scratch/larissa.reames/ideal_updrafts/WRF_C3500_25s/shallow_12.5/wrfout*"
fsW = glob.glob(string)
fsW.sort()

string = "/scratch/larissa.reames/ideal_updrafts/CM1_C3500_25s/shallow_12.5/*"
fsC = glob.glob(string)
fsC.sort()

z_interp = np.arange(10.0,24010.0,100.0)

filenameW = fsW[0]
fW = xr.open_dataset(filenameW ,engine='netcdf4')
p0W = np.array(fW.P[0,:,0,0]+fW.PB[0,:,0,0])
zW = (fW.PH[0,:,0,0]+fW.PHB[0,:,0,0])/9.81 - fW.HGT[0,0,0]
dz0W = zW[1:]-zW[:-1]
z0W = 0.5 * (zW[:-1]+zW[1:])
p0W = np.interp(z_interp,z0W,p0W)

p0F = np.array(fF.pres[0,::-1,0,0])
vap = np.array(fF.spfh[0,::-1,0,0])
p0F = p0F / (1 + vap * 0.622)
dz0F = -fF.delz[0,::-1,0,0]
z0F = np.array(-np.cumsum(fF.delz[0,::-1,0,0],axis=0))
p0F = np.interp(z_interp,z0F,p0F)


filename = fsC[0]
f = xr.open_dataset(filename ,engine='netcdf4')
p0C = np.array(f.prs0[0,:,0,0]+f.prspert[0,:,0,0])
zC = np.array(f.zh)*1000.0
p0C = np.interp(z_interp,zC,p0C)

cbar_set = False
# Create the figure
#fig,ax = plt.subplots(1,2,sharey='row')
#fig.set_size_inches(32,8)
fig = plt.figure(figsize=(16, 8))

grid = AxesGrid(fig, 111,
                nrows_ncols=(1, 3),
                axes_pad = 0.1,
                cbar_mode='single',
                cbar_location='bottom',
                cbar_pad=0.75,
                share_all=False,
                aspect=False
                )

#fig.savefig("/work/larissa.reames/w_cross_BOTH_C3500_y=%i_t=%i.png"%(cross_y,(t+1)*25),dpi=300)
ani = animation.FuncAnimation(fig,plot_cross,np.arange(24,72),interval=300,blit=False)
writer = animation.writers['ffmpeg'](fps=2)
ani.save('%s_cross_comparison_fixedrad.mp4'%(plot_var),writer=writer,dpi=300)
```
